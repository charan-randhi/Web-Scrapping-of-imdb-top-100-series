```python
# Import useful libraries and classes.
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
my_url_1=  "https://www.imdb.com/search/title/?title_type=tv_series&year=1996-01-01,2021-12-31&sort=num_votes,desc"
my_url_2= "https://www.imdb.com/search/title/?title_type=tv_series&year=1996-01-01,2021-12-31&sort=num_votes,desc&start=51&ref_=adv_nxt"
uClient1= uReq(my_url_1)
uClient2=uReq(my_url_2)
page_html1= uClient1.read()
page_html2=uClient2.read()
uClient1.close()
uClient2.close()
```

```python
page_html1
```

```python
page_soup1= soup(page_html1)
page_soup1
```

```python
page_soup2=soup(page_html2)
page_soup2
```

```python
#read class from web page.
containers1= page_soup1.findAll("div", {"class": "lister-item mode-advanced"})
print(len(containers1))
```

```python
containers2= page_soup2.findAll("div", {"class": "lister-item mode-advanced"})
print(len(containers1))
```

```python
filename= "imdb_s.csv"
f= open(filename, "w")

headers= "Name,Startyear,EndYear,Runtime,Genre1,rating,votes\n"
f.write(headers)

for container in containers1:
    name = container.img["alt"]
    year_mov= container.findAll("span", {"class": "lister-item-year"})
    start_year = year_mov[0].text.strip(')').strip('(').split('')[0]
    End_year = year_mov[0].text.strip(')').strip('(').split('')[1]
    runtime_mov= container.findAll("span", {"class": "runtime"})
    runtime = runtime_mov[0].text
    genre_mov= container.findAll("span", {"class": "genre"})
    genre = genre_mov[0].text[1:].strip().split(',')[0]
    imdb_rating = container.findAll('strong')
    rating = imdb_rating[0].text
    votes_mov= container.findAll("span", {"name": "nv"})
    votes = ''.join(votes_mov[0].text.strip().split(','))
    print(name + "," + start_year + ","+ End_year+","+ runtime +"," + genre + "," + rating + "," + votes +"\n")
    f.write(name + "," + start_year + ","+ End_year+","+ runtime +"," + genre + "," + rating + "," + votes +"\n")

for container in containers2:
    name = container.img["alt"]
    year_mov= container.findAll("span", {"class": "lister-item-year"})
    start_year = year_mov[0].text.strip(')').strip('(').split('')[0]
    End_year = year_mov[0].text.strip(')').strip('(').split('')[1]
    runtime_mov= container.findAll("span", {"class": "runtime"})
    runtime = runtime_mov[0].text
    genre_mov= container.findAll("span", {"class": "genre"})
    genre = genre_mov[0].text[1:].strip().split(',')[0]
    imdb_rating = container.findAll('strong')
    rating = imdb_rating[0].text
    votes_mov= container.findAll("span", {"name": "nv"})
    votes = ''.join(votes_mov[0].text.strip().split(','))
    
    print(name + "," + start_year + ","+ End_year+","+ runtime +"," + genre + "," + rating + "," + votes +"\n")
    f.write(name + "," + start_year + ","+ End_year+","+ runtime +"," + genre + "," + rating + "," + votes +"\n")

    
f.close()
```
